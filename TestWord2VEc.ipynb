```python
from bd_gensim.doc2vec_facade import Doc2VecFacade
import yaml
```

```python
config = yaml.safe_load(open("config.yml"))
data_dir = config['data_dir']
models_dir = config['models_dir']
```

```python
doc2VecFacade = Doc2VecFacade(models_dir)
```

```python
doc2VecFacade.load_models()
```

```python
model = doc2VecFacade.model
```

```python
cw_l = []
for word, vocab_obj in model.wv.vocab.items():
    cw_l.append((word, vocab_obj.count))
cw_s = sorted(cw_l, key=lambda x : x[1], reverse=True)
```
